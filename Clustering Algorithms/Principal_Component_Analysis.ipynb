**Principal Component Analysis (PCA)**

Principal Component Analysis (PCA) is a powerful unsupervised learning technique used for dimensionality reduction and data visualization. It transforms large sets of correlated features into a smaller set of uncorrelated variables known as principal components, which capture the maximum variance in the data. By projecting data onto these new axes of maximum variation, PCA helps uncover hidden patterns, relationships, and structures that might not be immediately visible in high-dimensional datasets.

We implement PCA to simplify complex datasets, reduce noise, and visualize data effectively without losing significant information. In this project, PCA is applied to the Breast Cancer dataset to understand how multiple medical features can be summarized into a few principal components. The reduced dimensions are then visualized to explore natural separations between benign and malignant tumors. This demonstrates how PCA can help in feature extraction, pattern recognition, and data preprocessing for other machine learning models.

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
sns.set(style="whitegrid")
```

```python
# Load and Explore Dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Exploratory Data Analysis (EDA)
df['target'].value_counts() #Check class balance
```

```python
# Feature Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.iloc[:,:-1].corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Distributions of new features
df.iloc[:, :5].hist(figsize=(10,6), bins=20)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Data Preprocessing
X = df.drop('target', axis=1)
y = df['target']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA(n_components=None)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Variance Analysis
explained_variance = pca.explained_variance_ratio_
cum_var = np.cumsum(explained_variance)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(cum_var)+1), cum_var, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA Components")
plt.grid(True)
plt.show()
```

```python
# Reduce Dimensions (2D & 3D Visualization)
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_2d[:,0], y=X_2d[:,1], hue=y, palette='coolwarm', s=60)
plt.title("PCA (2 Components) - Data Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
X_3d = pca_3d.fit_transform(X_scaled)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_3d[:,0], X_3d[:,1], X_3d[:,2], c=y, cmap='coolwarm', s=50)
ax.set_title("3D PCA Visualization")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```

```python
# Feature Contribution to Principal Components
loadings = pd.DataFrame(pca_2d.components_.T,
                        columns=['PC1', 'PC2'],
                        index=X.columns)
plt.figure(figsize=(10,6))
sns.heatmap(loadings, annot=True, cmap='YlGnBu')
plt.title("Feature Contribution to Principal Components")
plt.show()
```

```python
# Svae PCA Model
from joblib import dump
dump(pca_2d, "pca_breast_cancer.joblib")
dump(scaler, "scaler_breast_cancer.joblib")
print("Saved PCA and scaler models.")
```

**Conclusion:-**

The Principal Component Analysis (PCA) technique effectively reduced the high-dimensional Breast Cancer dataset into a smaller number of components while retaining most of the important information. Through dimensionality reduction, PCA simplified the dataset and revealed clear patterns between benign and malignant samples, which were visualized effectively in 2D and 3D plots.

This implementation demonstrated PCA’s strength in capturing maximum variance, eliminating redundancy, and enhancing data interpretability. It serves as a powerful preprocessing and visualization tool for unsupervised learning tasks, improving both model efficiency and understanding of the underlying data structure.
